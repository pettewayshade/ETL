```python
import os 
from config import key
import pandas as pd
import requests
import warnings
import pyodbc
from sqlalchemy import create_engine
from server import server, database, username, password

#ignore future warnings error
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
columns = ["population", "state"]
popData = pd.DataFrame(columns=columns)
```

```python
baseURL = "https://api.census.gov/data/2017/pep/population?get="
targetURL = (baseURL + "POP,GEONAME&for=state:*&key=" + key) 
resp = requests.get(targetURL).json()
resp
```

```python
count = 0
firstline = True

for item in resp:
    if firstline:
        firstline = False
        continue
        
    popData.set_value(count, "population", item[0])
    popData.set_value(count, "state", item[1])
    count +=1
    
popData.head(50)
```

```python
export_csv = popData.to_csv('datasets/populationData.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```
